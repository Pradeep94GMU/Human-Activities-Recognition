```python
#read each file and from prefix and create a lable as activity based on the filename
def read_and_label_files(directory, activity_prefix):
    data_list = []
    for filename in os.listdir(directory):
        if filename.startswith(activity_prefix) and filename.endswith('.csv'):
            file_path = os.path.join(directory, filename)
            data = pd.read_csv(file_path)
            data = smooth_data(data)  # Applying smoothing filter
            data['activity'] = activity_prefix  # Using the activity prefix as the label
            data_list.append(data)
    return pd.concat(data_list, ignore_index=True)

#smooth the data, applying low pass filter... (preproccesing steps)
def smooth_data(data):
    for column in ['x', 'y', 'z']:
        data[column] = savgol_filter(data[column], window_length=5, polyorder=2)
    return data


#this funtion will create a windows ..
def segment_into_windows(data, points_per_window):
    windows = []
    for start in range(0, len(data), points_per_window):
        end = start + points_per_window
        # Discarding the last window if it has fewer points than `points_per_window`
        if end <= len(data):
            window = data.iloc[start:end]
            windows.append(window)
    return windows

#labling after reading the filesss..
walking_data = read_and_label_files('/home/7a70dade-6117-4027-868d-b2f57326bc5b/dec', 'walking')
sitting_data = read_and_label_files('/home/7a70dade-6117-4027-868d-b2f57326bc5b/dec', 'sitting')
running_data = read_and_label_files('/home/7a70dade-6117-4027-868d-b2f57326bc5b/dec', 'running')
rotating_data = read_and_label_files('/home/7a70dade-6117-4027-868d-b2f57326bc5b/dec', 'rotating')
standing_data = read_and_label_files('/home/7a70dade-6117-4027-868d-b2f57326bc5b/dec', 'standing')
liftUp_data = read_and_label_files('/home/7a70dade-6117-4027-868d-b2f57326bc5b/dec', 'liftUp')
liftDown_data = read_and_label_files('/home/7a70dade-6117-4027-868d-b2f57326bc5b/dec', 'liftDown')
stairUp_data = read_and_label_files('/home/7a70dade-6117-4027-868d-b2f57326bc5b/dec', 'stairUp')
stairDown_data = read_and_label_files('/home/7a70dade-6117-4027-868d-b2f57326bc5b/dec', 'stairDown')



# Combine all data
all_data = pd.concat([walking_data, sitting_data,running_data,rotating_data,standing_data,liftUp_data, liftDown_data, stairUp_data, stairDown_data], ignore_index=True)
all_data.shape

# Segmenting into windows
points_per_window = 58 * 2  # 58 Hz(sampling freq rate * 2 seconds windows
windows = segment_into_windows(all_data, points_per_window)

# Now, `windows` is a list where each element is a DataFrame representing a 2-second window

#debug
print(f"Number of windows: {len(windows)}")
print(f"Number of Data's (Rows, Col): {all_data.shape}")


activity_counts = all_data['activity'].value_counts()
print(activity_counts)


def extract_features(window):
    features = {}
    
    if 'activity' in window.columns:
        features['activity'] = window['activity'].iloc[0]
    
    # Statistical features
    for axis in ['x', 'y', 'z']:
        data = window[axis]
        features[f'{axis}_mean'] = np.mean(data)
        features[f'{axis}_std'] = np.std(data)
        features[f'{axis}_var'] = np.var(data)
        features[f'{axis}_median'] = np.median(data)
        features[f'{axis}_skew'] = stats.skew(data)
        features[f'{axis}_kurtosis'] = stats.kurtosis(data)
    
    # Time-domain features
    for axis in ['x', 'y', 'z']:
        data = window[axis]
        features[f'{axis}_rms'] = np.sqrt(np.mean(data**2))
        features[f'{axis}_zero_crossing_rate'] = np.sum(np.diff(np.sign(data)) != 0) / len(data)
        features[f'{axis}_sma'] = np.sum(np.abs(data))
        features[f'{axis}_energy'] = np.sum(data**2)

        # Entropy (using histogram for probability distribution)
        hist, bin_edges = np.histogram(data, bins=10, density=True)
        hist = np.nan_to_num(hist, nan=0.0)  # Replace NaN with 0
        hist = hist / hist.sum()  # Normalize to create a probability distribution
        entropy = -np.sum(hist * np.log2(hist + 1e-9))  # Log of zero handled by adding a small value (1e-9)
        features[f'{axis}_entropy'] = entropy
    
    # Frequency-domain features (using FFT)
    for axis in ['x', 'y', 'z']:
        data = window[axis]
        fft_values = np.fft.fft(data)
        fft_values = np.abs(fft_values)
        dominant_frequency = np.argmax(fft_values)
        spectral_entropy = -np.sum((fft_values / np.sum(fft_values)) * np.log2(fft_values / np.sum(fft_values) + 1e-9))
        features[f'{axis}_dominant_frequency'] = dominant_frequency
        features[f'{axis}_spectral_entropy'] = spectral_entropy
    
    return features


window_features = [extract_features(window) for window in windows]

# Now, `window_features` is a list where each element is a dictionary containing features for a window

from sklearn.feature_selection import mutual_info_classif
import pandas as pd

# Convert the list of feature dictionaries to a DataFrame
feature_df = pd.DataFrame(window_features)

# Separating features and target variable
X = feature_df.drop('activity', axis=1)  # Features
y = feature_df['activity']  # Target variable (activity labels)


feature_df.head()
activity_counts = feature_df['activity'].value_counts()
print(activity_counts)


from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

#shuffle=True,stratify=y

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


# Model Training
rf_model = RandomForestClassifier(n_estimators=1500, random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Model Evaluation
print(f"Random forest Accuracy: {accuracy * 100:.2f}%")
print(classification_report(y_test, y_pred))

y_test.value_counts()

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Compute confusion matrix
confusion = confusion_matrix( y_test, y_pred)

# Define class/activity labels
class_labels = ['liftDown','liftup','rotating','running', 'sitting', 'walking','stairUp','standing', 'stairDown' ] 

plt.figure(figsize=(8, 6))
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Assuming you have multiple unseen datasets
unseen_data_files = ['UnseenSitting1.csv', 'UnseenRotating.csv', 'UnseenWalking1.csv']  # Add more file names as needed

# Initialize an empty list to store activity percentages
activity_percentages_list = []

# Loop through each unseen dataset
for data_file in unseen_data_files:
    # Load the unseen data (replace this with your data loading code)
    unseen_data = pd.read_csv(data_file)
    
    # Smooth and preprocess the data if needed
    # unseen_data = smooth_data(unseen_data)
    
    # Segment into windows and extract features
    unseen_windows = segment_into_windows(unseen_data, points_per_window=58*2)
    unseen_window_features = [extract_features(window) for window in unseen_windows]
    
    # Convert to DataFrame
    unseen_feature_df = pd.DataFrame(unseen_window_features)
    
    # Load the saved model
    rf_model = load('HAR1_trained_model.joblib')
    
    # Make predictions
    unseen_predictions = rf_model.predict(unseen_feature_df)
    
    # Calculate the count and percentage of each activity
    activity_counts = Counter(unseen_predictions)
    total_count = len(unseen_predictions)
    activity_percentages = {activity: count / total_count * 100 for activity, count in activity_counts.items()}
    
    # Append the activity percentages to the list
    activity_percentages_list.append(activity_percentages)

# Create a Pandas DataFrame from the list of activity percentages
df = pd.DataFrame(activity_percentages_list, index=[f'Dataset {i+1}' for i in range(len(unseen_data_files))])

# Transpose the DataFrame for better readability
df = df.transpose()

# Display the table
print(df)

# Optionally, save the table as a CSV file
df.to_csv('activity_percentages_table.csv', index=True)

# Optionally, you can also display the table as a LaTeX table for inclusion in presentations or reports
print(df.to_latex(index=True))
```
