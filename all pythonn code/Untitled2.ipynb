```python
import os
import pandas as pd
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
activity_data = pd.read_csv('gyrocsc/Accelerometer.csv')

# Plot the 'x' axis accelerometer data
plt.figure(figsize=(9, 5))
plt.plot(activity_data['seconds_elapsed'], activity_data['x'])
plt.title('Running Activity (X-axis)')  
plt.xlabel('Time (seconds)')
plt.ylabel('Accelerometer (m/s^2)')
plt.grid(True)
plt.show()
```
